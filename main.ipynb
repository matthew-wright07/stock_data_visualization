```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("stock_data.csv")

data = data.set_index("Index")

close = data.loc["HSI"]["Close"]

x = range(len(close))

plt.figure(figsize=(10,12))
plt.plot(x, close, marker='o', linestyle='-', color='blue')
plt.title("Close Prices for HSI")
plt.xlabel("Data Point")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```
